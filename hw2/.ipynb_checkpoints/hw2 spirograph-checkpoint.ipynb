```python
# setup
import numpy as np
import matplotlib.pyplot as plt

# https://www.wikiwand.com/en/Spirograph for formula
rho =3
r = 98
R = 100
l = rho / r
k = r / R

thetas = np.linspace(0,np.pi*2*r,num=1000)
```

```python
# using imperative
X = []
Y = []
for i in range(0,len(thetas)):
    X.append(R*((1-k)*np.cos(thetas[i])+l*k*np.cos((1-k)/k*thetas[i])))

for i in range(0,len(thetas)):
    Y.append(R*((1-k)*np.sin(thetas[i])-l*k*np.sin((1-k)/k*thetas[i])))

plt.plot(X,Y)
plt.axis('equal')
plt.show()
```

```python
# using functional
def calcX(angle):
    return (R*((1-k)*np.cos(angle)+l*k*np.cos((1-k)/k*angle)))

def calcY(angle):
    return (R*((1-k)*np.sin(angle)-l*k*np.sin((1-k)/k*angle)))


Xs = list(map(calcX,thetas))
Ys = list(map(calcY,thetas))

plt.plot(Xs,Ys)
plt.axis('equal')
plt.show()
```
