```python
# setup
import numpy as np
import matplotlib.pyplot as plt

# https://www.wikiwand.com/en/Spirograph for mathematical formula
thetas = np.arange(-np.pi*100,np.pi*100,5)
rho =10
r = 80
R = 100
l = rho / r
k = r / R

X = []
Y = []
for i in range(0,len(thetas)):
    X.append(R*((1-k)*np.cos(thetas[i])+l*k*np.cos((1-k)/k*thetas[i])))

for i in range(0,len(thetas)):
    Y.append(R*((1-k)*np.sin(thetas[i])-l*k*np.sin((1-k)/k*thetas[i])))

plt.plot(X,Y)
plt.axis('equal')
plt.show()
```
